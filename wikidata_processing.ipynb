```python
import pandas as pd

df = pd.read_csv('Datasets/14_graph.tsv', sep='\t', header=None)

df_0 = df.loc[df.iloc[:, 0].str.contains('/Q') == True]
# Q_uniques = df_0.iloc[:, 0].unique()
Q_uniques = df_0.iloc[:, 0].apply(lambda x: x[32:-1]).unique()
```

```python
import time
import pickle
start = time.time()
current_time = start

Q2name = dict()
name2Q = dict()

site = pywikibot.Site("wikidata", "wikidata")
repo = site.data_repository()

invalid_count = 0
invalid_Qs = []

for i, Q_entity in enumerate(Q_uniques):
    # Q_entity = Q_wiki.removeprefix('<http://www.wikidata.org/entity/').removesuffix('>')
    # Q_entity = Q_wiki[32:-1]
    item = pywikibot.ItemPage(repo, Q_entity)
    if item.isRedirectPage():
        item = item.getRedirectTarget()
    if not item.exists():
        invalid_count += 1
        invalid_Qs.append(Q_uniques)
        print('Invalid Q: {}'.format(Q_entity))
        continue
    item_dict = item.get()
    labels = item_dict['labels']
    if 'en' not in list(labels.keys()):
        invalid_count += 1
        invalid_Qs.append(Q_uniques)
        print('Invalid Q: {}'.format(Q_entity))
        continue
    name = labels['en']
    Q2name.update({Q_entity: name})
    name2Q.update({name: Q_entity})
    if i % 100 == 0:
        print('{} / {} {}: {} time: {}/{}'.format(i, len(Q_uniques), Q_entity,name, time.time() - current_time, time.time() - start))
    # if i == 1499:
    #     # store data
    #     with open('1500_Q2name.pickle', 'wb') as handle:
    #         pickle.dump(Q2name, handle, protocol=pickle.HIGHEST_PROTOCOL)
    #     with open('1500_name2Q.pickle', 'wb') as handle:
    #         pickle.dump(name2Q, handle, protocol=pickle.HIGHEST_PROTOCOL)
    if (i+1) % 10000 == 0:
        #store data
        with open('{}_Q2name.pickle'.format((i+1)//10000), 'wb') as handle:
            pickle.dump(Q2name, handle, protocol=pickle.HIGHEST_PROTOCOL)
        with open('{}_name2Q.pickle'.format((i+1)//10000), 'wb') as handle:
            pickle.dump(name2Q, handle, protocol=pickle.HIGHEST_PROTOCOL)
        Q2name = {}
        name2Q = {}
    current_time = time.time()

#store data
with open('_end_Q2name.pickle', 'wb') as handle:
    pickle.dump(Q2name, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('_end_name2Q.pickle', 'wb') as handle:
    pickle.dump(name2Q, handle, protocol=pickle.HIGHEST_PROTOCOL)
print('Work done')
```
