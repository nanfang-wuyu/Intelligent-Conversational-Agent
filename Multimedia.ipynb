```python
import ijson
```

```python
id2Code = dict()
total_id_cnt = len(id_uni)
current_cnt = 0
with open("Datasets/ImageData/images.json", "rb") as f:
    for record in ijson.items(f, "item"):
        movies = record["movie"]
        cast_members = record["cast"]
        magic_code = record["img"]
        # print(movies, cast_members, magic_code)
        ids = movies
        ids.extend(cast_members)
        for id in ids:
            if (id not in id2Code) and (id in id_uni):
                id2Code[id] = magic_code
                current_cnt += 1
        if current_cnt >= total_id_cnt:
            break
        if current_cnt % 1000 == 0:
            print('{}/{}'.format(current_cnt,total_id_cnt))
print(len(id2Code))
        
        #     user_to_repos[user] = set()
        # user_to_repos[user].add(repo)
```

```python
import pickle
with open('Datasets/ImageData/id2Code.pickle', 'wb') as handle:
    pickle.dump(id2Code, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('Datasets/ImageData/id2Code.pickle', 'rb') as handle:
    di = pickle.load(handle)


```
